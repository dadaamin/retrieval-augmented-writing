```python
from llama_index.core.readers import SimpleDirectoryReader
```

```python
from raw.loader import PDFReaderPlus, TesseractReader
```

```python
pdf_reader = PDFReaderPlus(language='deu')
image_reader = TesseractReader()

custom_readers = {
    '.pdf': pdf_reader,
    '.png': image_reader,
    '.jpg': image_reader,
    '.jpeg': image_reader
}

reader = SimpleDirectoryReader('../test-data/', filename_as_id=True, file_extractor=custom_readers)
docs = reader.load_data()
docs
```
